1. Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

2. Creating DataFrame and Saving to CSV

```python
num_points = 50
x = np.random.randint(0,100,num_points)
y = 2 * x + np.random.normal(0,10,num_points)
data = pd.DataFrame({'X':x,'Y':y})

data.to_csv('data.csv',index=False)

print(data.head())
```

3. Visualizing the Data

```python
plt.scatter(data['X'],data['Y'])
plt.xlabel('Independent Variable')
plt.ylabel('Dependent Variable')
plt.title('Scatter Plot of Data')
plt.show()
```

4. Preparing for Data Modeling


```python
X = data[['X']]
Y = data['Y']
```

5. Fitting the Linear Regression Model

```python
model = LinearRegression()
model.fit(X,Y)
```

6. Getting Model Coefficients

```python
slope = model.coef_[0]
intercept = model.intercept_
print("Slope: ",slope)
print("Intercept: ",intercept)
```

7. Visualizing the Linear Regression Line


```python
plt.scatter(X,Y, color='teal')
plt.plot(X,model.predict(X),color='red')
plt.xlabel('Independent Variable')
plt.ylabel('Dependent Variable')
plt.title('Linear Regerssion')
plt.show()
```

8. Making Predictions

```python
new_X = [[np.random.randint(100,200)]]
predicted_Y = model.predict(new_X)
print("New X: ", new_X)
print("Predicted Y for New X: ", predicted_Y)
```
